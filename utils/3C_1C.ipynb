```python
import numpy as np
import os
import cv2
path=os.getcwd()
```

```python
ls=os.listdir(path)
lst=[(i) for i in ls if i.endswith('.png')]
```

```python
lst.sort
lst
```

```python
cv2.imread(lst[1]).shape
len(lst)
```

```python
for i in lst:
    img=cv2.imread(i)
    img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    cv2.imwrite(i,img)
    
    
```
